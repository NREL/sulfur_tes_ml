```python
import os
import pandas as pd
import glob

import xgboost
print(xgboost.__version__)

from sklearn.datasets import make_classification
from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import KFold
from xgboost import XGBRegressor

from sklearn.metrics import r2_score

%config InlineBackend.figure_formatrmat = 'retina'
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
data_dir = "../data/Air_Models/"

scenario_index = pd.DataFrame({"filepath": glob.glob(os.path.join(data_dir, "IG_*_*.csv"))})
scenario_index
```

```python
cv = RepeatedKFold(n_splits=5, n_repeats=1, random_state=5)

train_index, test_index  = next(cv.split(scenario_index.index))

#train_index, test_index = cv.split(range(50))
 
print("TRAIN:", train_index, "TEST:", test_index)
print("TRAIN count:", len(train_index), "TEST count:", len(test_index))
```

```python
def load_data(scenario_index, selected_index):
    """ Load data from files in scenario_index with indices matching ones in selected_index"""
    
    df_arr = []
    for f in scenario_index.loc[selected_index].filepath:
        Tw = float(f.split("/")[-1].split("_")[1])
        Ti = float(f.split("/")[-1].split("_")[2].replace(".csv", ""))

        f_df = pd.read_csv(f, skiprows=12)
        f_df["Tw"] = Tw
        f_df["Ti"] = Ti
        df_arr.append(f_df)
    
    combined_df = pd.concat(df_arr)
    return combined_df

train_df = load_data(scenario_index, train_index)
train_df
```

```python
X = train_df[["Time", "Tw", "Ti"]].to_numpy()
X.shape
```

```python
y = train_df[["Tc"]].to_numpy().reshape(-1,)
print(y.shape)
y
```

```python
test_df = load_data(scenario_index, test_index)
X_test = test_df[["Time", "Tw", "Ti"]].to_numpy()
y_test = test_df[["Tc"]].to_numpy().reshape(-1,)
print(X_test.shape, y_test.shape)
```

```python
n_estimators = 1
model = XGBRegressor(n_estimators=n_estimators, colsample_bylevel=.75)
eval_set = [(X_test, y_test)]
model.fit(X, y, eval_metric="rmse", early_stopping_rounds=50, eval_set=eval_set, verbose=True)
```

```python
model = XGBRegressor(n_estimators=n_estimators, colsample_bylevel=.75)

min_estimators = 1
max_estimators = 10

r2_score_hist = list()

for i in range(min_estimators, max_estimators + 1, 1):
    model.set_params(n_estimators=i)
    model.fit(X, y)

    # Record the R2 for each `n_estimators=i` setting.
    y_hat = model.predict(X_test)
    r2 = r2_score(y_test,y_hat)
    #mse = mean_squared_error(y_test,y_hat)
    #mserror_rate.append((i, mse))
    r2_score_hist.append((i, r2))
    print(i, r2)
    
min_estimators = 11
max_estimators = 31

for i in range(min_estimators, max_estimators + 1, 2):
    model.set_params(n_estimators=i)
    model.fit(X, y)

    # Record the R2 for each `n_estimators=i` setting.
    y_hat = model.predict(X_test)
    r2 = r2_score(y_test,y_hat)
    #mse = mean_squared_error(y_test,y_hat)
    #mserror_rate.append((i, mse))
    r2_score_hist.append((i, r2))
    print(i, r2)

min_estimators = 32
max_estimators = 102

for i in range(min_estimators, max_estimators + 1, 10):
    model.set_params(n_estimators=i)
    model.fit(X, y)

    # Record the R2 for each `n_estimators=i` setting.
    y_hat = model.predict(X_test)
    r2 = r2_score(y_test,y_hat)
    #mse = mean_squared_error(y_test,y_hat)
    #mserror_rate.append((i, mse))
    r2_score_hist.append((i, r2))
    print(i, r2)

min_estimators = 102
max_estimators = 302

for i in range(min_estimators, max_estimators + 1, 20):
    model.set_params(n_estimators=i)
    model.fit(X, y)

    # Record the R2 for each `n_estimators=i` setting.
    y_hat = model.predict(X_test)
    r2 = r2_score(y_test,y_hat)
    #mse = mean_squared_error(y_test,y_hat)
    #mserror_rate.append((i, mse))
    r2_score_hist.append((i, r2))
    print(i, r2)
```

```python
Xs = []
Ys = []

for model_err in r2_score_hist:
    Xs.append(model_err[0])
    Ys.append(model_err[1])
    #plt.plot(xs, ys)
    #print(model_err)

fig, ax = plt.subplots(figsize=(12,8), dpi= 200, facecolor='w', edgecolor='k')
ax.plot(Xs,Ys)
#ax.set_yscale('log')
plt.xlim(30, 300)
plt.ylim(0.996, .997)
#plt.ticklabel_format(style='plain')
plt.xlabel("n_estimators")
plt.ylabel("R2")
plt.title("non-Recurrent XGBoost Model: R2 Score for Test Data")
plt.show()
```
