```python
import os
import pandas as pd
import glob

data_dir = "../data/Air_Models/"
df_arr = []

for f in glob.glob(os.path.join(data_dir, "IG_*_*.csv")):
    Tw = float(f.split("/")[-1].split("_")[1])
    Ti = float(f.split("/")[-1].split("_")[2].replace(".csv", ""))
    
    f_df = pd.read_csv(f, skiprows=12)
    f_df["Tw"] = Tw
    f_df["Ti"] = Ti
    df_arr.append(f_df)
    
    #print(f_df.columns)
    
    #display(f_df)
    #break
    #print(len(f_df))
    
    #break
    
combined_df = pd.concat(df_arr)
combined_df
```

```python
# Source of this example: https://machinelearningmastery.com/extreme-gradient-boosting-ensemble-in-python/

from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import KFold
from xgboost import XGBRegressor
# define dataset
#X, y = make_classification(n_samples=1000, n_features=20, n_informative=15, n_redundant=5, random_state=7)
# define the model

model = XGBRegressor()
# evaluate the model


# Original code (for classification):
#cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
# It causes an error:
# ValueError: Supported target types are: ('binary', 'multiclass'). Got 'continuous' instead.
# From: https://stackoverflow.com/questions/48508036/sklearn-stratifiedkfold-valueerror-supported-target-types-are-binary-mul
# "If your target variable is continuous then use simple KFold cross validation instead of StratifiedKFold."

cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
#print(cv)

# Debug info
# for train_index, test_index in cv.split(X):
#     print("TRAIN:", train_index, "TEST:", test_index)
#     print("TRAIN:", len(train_index), "TEST:", len(test_index))
#     X_train, X_test = X[train_index], X[test_index]
#     y_train, y_test = y[train_index], y[test_index]

n_scores = cross_val_score(model, X, y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1, error_score='raise')
# report performance
print('MAE: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```
